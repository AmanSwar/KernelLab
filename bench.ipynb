```python
import matplotlib.pyplot as plt
```

```python
numbers = [23, 45, 55, 12, 38, 62, 41]
labels = [
    "Category A",
    "Category B",
    "Category C",
    "Category D",
    "Category E",
    "Category F",
    "Category G",
]
```

```python
if len(numbers) != len(labels):
    # If not, create simple default labels based on index
    labels = [f"Point {i+1}" for i in range(len(numbers))]
```

```python
plt.figure(figsize=(10, 6))

# The plt.bar() function creates the bar plot.
# It takes the x-axis positions (labels) and y-axis values (numbers).
plt.bar(labels, numbers, color="darkgreen")

# 4. Add titles and labels for clarity
plt.title("Bar Graph of Numbers")
plt.xlabel("Categories")
plt.ylabel("Values")

# 5. Display the graph
# plt.show() renders and opens a window with your graph.
plt.show()
```
